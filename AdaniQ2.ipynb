#### •	Removing the 'Total of' in every row of the 'Name of Thermal Power Station/ Performance of Utility#' column 

```python
is_null_first_two = report.iloc[:,0:2].isna().all(axis=1)  # Check if all values in first two columns are NA
report = report[~is_null_first_two]  # Select rows where ~is_null_first_two is True (not all NA)
len(report)
```

#### Dropping Report Summary Table 

```python
# Reset the index
report = report.reset_index(drop=True)  # Optional: drop the old index column


# Find the index of the last occurrence of 'REPORT SUMMARY' (using .idxmax)
value_to_find= 'REPORT SUMMARY'
index = report['Sl No. / State'].eq(value_to_find).idxmax()  # Might raise IndexError if not found

# Print the index (if found)
if index is not None:
    print(f"Index of the last '{value_to_find}':", index)
else:
    print(f"'{value_to_find}' not found in 'col1'")
```

```python


# Drop rows (including the row at the index) from Report Summary
report=report.iloc[:209]  
```

#### Fetching rows with Plant Names

```python


# Check if any row in the column starts with r'^[ABCD]\.'
is_starts_with_A = list(report['Sl No. / State'].str.startswith('A.').fillna(False)).index(True)
is_starts_with_B = list(report['Sl No. / State'].str.startswith('B.').fillna(False)).index(True)
is_starts_with_C = list(report['Sl No. / State'].str.startswith('C.').fillna(False)).index(True)
is_starts_with_D= list(report['Sl No. / State'].str.startswith('D.').fillna(False)).index(True)

indices_to_drop=[is_starts_with_A,is_starts_with_B,is_starts_with_C,is_starts_with_D]

indices_to_drop
```

```python
print(report['Sl No. / State'].iloc[1])
print(report['Sl No. / State'].iloc[175])
print(report['Sl No. / State'].iloc[193])
print(report['Sl No. / State'].iloc[200])
```

```python
#Column creation
col_vals=[report['Sl No. / State'].iloc[1] for _ in range(175)]
col_vals.extend([report['Sl No. / State'].iloc[175] for _ in range(175,193)])
col_vals.extend([report['Sl No. / State'].iloc[193] for _ in range(193,200)])
col_vals.extend([report['Sl No. / State'].iloc[200] for _ in range(200,len(report))])
```

```python
col_vals[208]
```

```python
report['Plant Name']=col_vals
```

```python
report.drop(indices_to_drop,inplace=True)
```

```python
#Dropping blank row in header
report.drop([0],inplace=True)
# Reset the index
report = report.reset_index(drop=True)  # Optional: drop the old index column
```

```python
report.head(3)
```

```python
report.info()
```

```python
#Changing 'Sl No. / State' column to have consistent categorical values
alpha='a'
for i in range(0,len(report['Sl No. / State'])):
     try:
         if(0<=int(report['Sl No. / State'][i])<=9999):
            report['Sl No. / State'][i]=alpha            
     except:
            alpha=report['Sl No. / State'][i]
```

```python
len(report)
```

```python
#Dropping extra rows that only specify state names
is_null_skip_first_last = report.iloc[:,1:-1].isna().all(axis=1)
report = report[~is_null_skip_first_last]  # Select rows where ~is_null_first_two is True (not all NA)
len(report)
```

#### Creating CSV

```python
report.to_csv('coal_report.csv',encoding='utf-8')  # Set index=False to avoid index column
```

#### Aggregating Averages ; Totals and averages calculated in csv file Excel

```python
report['Tentative PLF current month upto date'] = pd.to_numeric(report['Tentative PLF current month upto date'], errors='coerce')  # Replace NaN with NaN

Tentative_PLF_current_month_upto_date_Average = report.groupby('Sl No. / State')['Tentative PLF current month upto date'].mean()
Tentative_PLF_current_month_upto_date_Average.mean()
```

```python
report['Actual Stock as % of Normative  Stock']=pd.to_numeric(report['Actual Stock as % of Normative  Stock'],errors='coerce')
Actual_Stock_of_Normative_Stock_Average=report.groupby('Sl No. / State')['Actual Stock as % of Normative  Stock'].mean()
Actual_Stock_of_Normative_Stock_Average.mean()
```
